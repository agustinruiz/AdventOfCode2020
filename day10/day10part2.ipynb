## Calculamos la complejidad:
### Siendo: - `m = len(adapters_list)`
### Para el caso de la complejidad de tiempo:  
#### En este caso solo voy a tener que iterar por toda la tabla.
#### Por lo tanto la complejidad de tiempo sera aproximadamente `O(m)`

### Lo mismo pasa para la complejidad de espacio. solo tengo una tabla de largo m.
#### => la complejidad de espacio es aproximadamente `O(m)`


### Ahora si voy a la resolucion en codigo

````python
def differents_arranges(adapters_list):
    table = [0] * len(adapters_list)
    table[0] = 1
    
    for i in range(len(adapters_list)):
        if (i+1 < len(adapters_list)) and ((adapters_list[i+1]-adapters_list[i]) <= 3):
            table[i+1] += table[i]
        if (i+2 < len(adapters_list)) and ((adapters_list[i+2]-adapters_list[i]) <= 3):
            table[i+2] += table[i]
        if (i+3 < len(adapters_list)) and ((adapters_list[i+3]-adapters_list[i]) <= 3):
            table[i+3] += table[i]

    return table[-1]
````
